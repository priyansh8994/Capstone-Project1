```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Updated Sample DataFrame with odd-numbered values
data = {
    'Hotel Name': [1199, 801, 601, 299, 201],
    'City': [1, 1, 1, 1, 1],
    'Number of Reviews': [9999, 4501, 5499, 799, 6499],
    'Free WiFi': [1, 1, 1, 1, 1],
    'Family rooms': [1, 1, 1, 1, 1],
    'No-smoking rooms': [0, 1, 1, 1, 1],
    'Restaurant': [1, 0, 0, 0, 1],
    'Bar': [1, 0, 0, 0, 0],
    'Heating': [1, 0, 0, 0, 0],
    'Lift': [1, 0, 0, 0, 0],
    'Breakfast Cuisine 1': [1, 1, 1, 1, 1],
    'Breakfast Cuisine 2': [2, 2, 2, 2, 2],
    'Breakfast Cuisine 3': [3, 9, 5, 1, 9],
    'Room Type': [7, 5, 3, 5, 5],
    'Single Bed': [1, 1, 1, 1, 1],
    'Double Bed': [0, 0, 0, 0, 0],
    'Review Score': ['Superb', 'Very good', 'Good', 'Pleasant', 'Superb']
}

# DataFrame from the updated sample data
df = pd.DataFrame(data)

def train_and_predict(df, target_column='Review Score', random_state=8):
    X = df.drop([target_column], axis=1)
    y = df[target_column]

    best_model = RandomForestClassifier(random_state=random_state)
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)
    best_model.fit(X, y_encoded)

    sample = df.sample(n=5)
    sample_X = sample.drop([target_column], axis=1)
    sample_predictions = best_model.predict(sample_X)
    sample_predictions = label_encoder.inverse_transform(sample_predictions)

    return sample_predictions

sample_predictions = train_and_predict(df)
print("Sample Predictions:")
print(sample_predictions)
```
